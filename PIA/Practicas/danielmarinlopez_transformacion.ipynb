# Práctica 3: Data Wrangling con Pandas (III. Transformación)

En esta fase de transformación se nos pide crear nuevas columnas y eliminar otras en su lugar.

Cargamos el csv y lo limpiamos.

```python
import pandas as pd

data = pd.read_csv('/content/MailCustomer.csv', index_col = 'CustomerKey', sep = ';', encoding = 'ISO-8859 -1')

data
```

Lo primero será crear tres columnas que recogan el día, mes y año de nacimiento, y borramos la columna de la fecha de nacimiento. Para ello haremos el siguiente código:

```python
data['BirthDate'] = pd.to_datetime(data['BirthDate'])  # Aseguramos que es datetime

data['BirthDateDay'] = data['BirthDate'].dt.day
data['BirthDateMonth'] = data['BirthDate'].dt.month
data['BirthDateYear'] = data['BirthDate'].dt.year

del data['BirthDate'] # Borramos la columna de manera directa

data
```

Si comprobamos con la salida anterior donde está la columna `BirthDate` podemos ver que la información se ha guardado correctamente.

Lo siguiente será hacer una columna con una categoría para la edad, por lo que primero haremos una gráfica para ver como se distribuyen las edades. Para ello usaremos `matplotlib`.

```python
import matplotlib.pyplot as plt

plt.hist(data['Age'], bins=20)
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.title('Distribución de Edades')
```

Podemos observar claramente que la gráfica muestra que todas las edades son superiores a los 40 años. Por lo que podemos crear las siguientes categorías:

* [40 - 70]: Adulto (Adult)
* [70 - 110]: Mayor (Older)

Para agregar la columna `AgeRange` con los correspondientes rangos podemos hacerlo de la siguiente forma:

```python
# Definir los rangos y las etiquetas correspondientes
bins = [40, 50, 60, 70, 80, 90, float('inf')]
labels = ['40-50', '51-60', '61-70', '71-80', '81-90', '+90']

# Crear una nueva columna con los rangos
data['AgeRange'] = pd.cut(data['Age'], bins=bins, labels=labels, right=True)
data
```

Una vez creada la columna, podemos hacer otra gráfica que muestre el número de clientes que pertenecen a cada rango.

```python
# Contar los valores por rango
age_range_counts = data['AgeRange'].value_counts().sort_index()

# Crear el gráfico
plt.bar(age_range_counts.index.astype(str), age_range_counts.values, color='skyblue', edgecolor='black')
plt.title('Distribución de Rangos de Edad')
plt.xlabel('Rango de Edad')
plt.ylabel('Cantidad')
plt.xticks(rotation=45)  # Rotar las etiquetas si es necesario
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Vemos que hay una gran cantidad de personas adultas frente a la cantidad de personas mayores.
