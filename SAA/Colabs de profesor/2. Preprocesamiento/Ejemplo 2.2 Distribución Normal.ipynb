Calculemos la F.D $\Phi(z)$

```python
# LIBRERÍAS EMPLEADAS EN EL EJEMPLO
# Tratamiento de datos
# ==============================================================================
import scipy.stats as st

z = 1.96
p_value = st.norm.cdf(z) # Phi(z)
p_value
```

Ahora calculemos el valor inverso de la F.D., $z=\Phi^{-1}(p_{\hbox{value}})$

```python
z = st.norm.ppf(p_value) # Phi^{-1}(p_value)
z
```

Calculemos probabilidades con una
distribución normal estándar $Z\sim N(0,1)$, mediante la relación $P(a<Z< b)=\Phi(b)-\Phi(a)$

```python
# P(-0.34 < Z < 1.4)
st.norm.cdf(1.4) - st.norm.cdf(-0.34)
```

Calculemos probabilidades con una variable aleatoria $X\sim N(10,5)$. Calculemos $P(4.5 < X<6)$

```python
# Tipificamos los valores
mu = 10
sigma = 5
a = (4.5-mu)/sigma
b = (6-mu)/sigma
P = st.norm.cdf(b) - st.norm.cdf(a)
P
```
