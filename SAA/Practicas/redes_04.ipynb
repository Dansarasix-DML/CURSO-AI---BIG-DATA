# 🎯 Objetivo
Entrenar una red neuronal multicapa para clasificar puntos en tres clases diferentes. Se trabajará con un conjunto de datos sintético generado con make_classification.

# 📘 Contexto
Vas a usar make_classification para generar un conjunto de datos de 3 clases. El objetivo es que el modelo aprenda a distinguir entre estas clases mediante una red neuronal con una sola capa oculta.

# 🧠 Lo que vas a hacer
1. Generar datos con 3 clases usando make_classification.

2. Visualizar los datos en 2D.

3. Entrenar un MLPClassifier.

4. Evaluar su precisión.

5. Visualizar la predicción y la frontera de decisión.

# 🧾 Detalles del modelo
* MLP con una capa oculta de 10 neuronas.

* Función de activación: ReLU.

* Optimización: Adam.

* Salida automáticamente adaptada a 3 clases (Softmax interno en MLPClassifier).
