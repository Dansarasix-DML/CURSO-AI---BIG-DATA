# 🎯 Objetivo
Entrenar un modelo MLPClassifier (red neuronal multicapa) para clasificar puntos en dos clases usando un conjunto de datos sintético. Observar cómo la red aprende una frontera de decisión no lineal.

# 📘 Contexto
Vas a generar datos con make_moons, una función que crea dos medias lunas entrelazadas. Este problema no es linealmente separable, por lo que un MLP con al menos una capa oculta es necesario.

# 🧠 Lo que vas a hacer
1. Generar datos sintéticos con make_moons.

2. Visualizar los datos.

3. Entrenar una red neuronal con MLPClassifier.

4. Evaluar su rendimiento.

5. Visualizar la frontera de decisión aprendida con con np.meshgrid.

# 📌 Detalles del modelo
* Red neuronal con 1 capa oculta de 10 neuronas.

* Activación: ReLU.

* Optimizador: Adam.

* Ideal para mostrar cómo las redes aprenden fronteras no lineales.
