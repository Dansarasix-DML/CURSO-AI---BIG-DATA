# 🎯 Objetivo
Entrenar un modelo de red neuronal (perceptrón multicapa) para aprender una función continua. Usaremos el modelo MLPRegressor de Scikit-Learn para aproximar una función cuadrática simple.

# 📘 Contexto
Supongamos que tenemos un conjunto de datos generado con la siguiente función:

y = x^2 + ruido

El objetivo es que el modelo aprenda a predecir y a partir de x, es decir, que pueda ajustar una curva a los datos.

# 🧠 ¿Qué es una red neuronal en este contexto?
Una red neuronal es un modelo que puede aproximar funciones complejas, combinando muchas operaciones simples (como sumas, multiplicaciones y activaciones). Está compuesta por capas de neuronas artificiales.

En este ejercicio vamos a usar:

* ✅ Una capa oculta con 10 neuronas.

* 🔥 Función de activación ReLU: una función muy usada que transforma su entrada en 0 si es negativa, o la deja igual si es positiva. Ayuda a introducir no linealidad en la red.

* ⚙️ Optimizador Adam: un algoritmo de entrenamiento que ajusta los pesos de la red automáticamente, combinando lo mejor del descenso de gradiente clásico y adaptativo. Es rápido, robusto y ampliamente utilizado.

# 📝 Instrucciones

1. Genera un conjunto de datos con valores de x entre -1 y 1, y y = x² + ruido.

2. Divide los datos en entrenamiento y prueba.

3. Entrena un MLPRegressor con:

  * hidden_layer_sizes=(10,)

  * activation='relu'

  * solver='adam'

4. Evalúa el rendimiento del modelo usando el error cuadrático medio.

5. Visualiza la curva aprendida.
