## 🎯 Objetivo
En este ejercicio trabajarás con el clásico conjunto de datos Iris, y entrenarás un modelo de Perceptrón para detectar si una flor es de la especie Setosa (una de las tres clases de flor del conjunto). Explorarás cómo influye el uso de distintas características en el rendimiento del modelo.

## 📘 Contexto
El dataset Iris contiene 150 muestras de flores, divididas en tres especies:

* Setosa

* Versicolor

* Virginica

Cada muestra contiene 4 características numéricas:

* Largo del sépalo (cm)

* Ancho del sépalo (cm)

* Largo del pétalo (cm)

* Ancho del pétalo (cm)

## 🧠 Lo que vas a hacer

1. Convertir el problema en binario: clasificar si la flor es o no Setosa.

2. Entrenar un modelo de Perceptrón para esta tarea.

3. Evaluar el rendimiento del modelo en dos escenarios:

  * Usando solo las características del pétalo (características 3 y 4).

  * Usando todas las características disponibles (1 a 4).

4. Comparar los resultados obtenidos y reflexionar sobre el impacto del conjunto de características.

## 📝 Instrucciones

1. Carga el dataset Iris desde sklearn.datasets.

2. Crea una variable objetivo binaria: 1 si la flor es Setosa, 0 en caso contrario.

3. Selecciona el subconjunto de características según el caso:

  * Caso 1: Solo columnas 2 y 3 (largo y ancho del pétalo).

  * Caso 2: Todas las columnas.

4. Divide el conjunto de datos en entrenamiento y prueba (ej. 75%/25%).

5. Entrena un clasificador Perceptron usando sklearn.linear_model.

6. Evalúa la precisión del modelo en el conjunto de prueba.

7. Compara los resultados y responde las preguntas de reflexión.

##💬 Preguntas de reflexión
* ¿Cuál modelo tuvo mejor rendimiento?

* ¿Es suficiente con las características del pétalo para detectar la clase Setosa?

* ¿Crees que el modelo funcionaría igual de bien para detectar Versicolor o Virginica? ¿Por qué?
